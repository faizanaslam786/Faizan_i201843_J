```python
import pandas as pd
import numpy as np
```

### Hypothesis

```python
def hypothesis(X,thetas):
                               #h(Θ)=ΘX+Θ0
    return np.dot(X,thetas.T)     # return Y_prdicted
```

### Loss_function

```python
def loss(y_actual,y_pred):
    return np.mean((y_actual-y_pred)**2)
```

### Derivative_loss_func

```python
def der_loss(X,Y,thetas):
                                      # -2/n * np.dot(diff.T,X)
    h=np.dot(X,thetas.T)
    diff=Y-h
    r=np.dot(diff.T,X)
    
    return (-2*(r/len(X)))
```

```python
def derivative_loss(diff,X):
    
    r=np.dot(diff.T,X)
    return (-2*(r/len(X)))
```

### Gradieant Descent

```python
def grad_desc(old_thet,part_thet,alpha):
    return (old_thet-(alpha*part_thet))
```

## Task1 Implemeting on data set

```python
df=pd.read_csv(f'advertising.csv')
df
```

```python
extr=np.array([[1]]*len(df))
x=np.array(df.iloc[: , :-1])
x=np.append(x, extr,axis = 1) 
x
```

```python
y=np.array(df.iloc[:,-1:])

# from numpy import random
# thet=random.randint(100, size=(4))
thet=np.array([90,50,20,10])

print(x.shape)
print(y.shape)
```

```python
h=hypothesis(x,thet)   # hypothesis call
# print ('prdicted Y:\n',h)

los=loss(y,h)           # loss call
print('\nLoss :',los)

par_thet=derivative_loss(y-h,x) # derivative loss call -----------> function will result new partial thetas
# print('Partial thetas',par_thet)

new_thet=grad_desc(thet,par_thet,0.00003)
# print('updated thetas :',new_thet)

##########################################################################################################3

h=hypothesis(x,new_thet)         # new hypothesis call after updated thetas
# print ('\nprdicted Y:\n',h)

los=loss(y,h)                       #   decreased    loss call
print('Loss :',los)
```

```python
                             # Running 100000 iteration for decresing the loss 
for i in range(10000):
    
    par_thet=der_loss(x,y,new_thet)                     # partial derivative loss call
    new_thet=grad_desc(new_thet,par_thet,0.00003)        # upgrading thetas gradient descent function call
    h=hypothesis(x,new_thet)   # hypothesis call            # new preducted values
    los=loss(y,h)           # loss call                 # losss call
    print('Loss :',los)
```

```python
                             # Running 45 iteration for decresing loss to minima
for i in range(45):
    
    par_thet=der_loss(x,y,new_thet)                     # partial derivative loss call
    new_thet=grad_desc(new_thet,par_thet,0.003)        # upgrading thetas gradient descent function call
    h=hypothesis(x,new_thet)   # hypothesis call            # new preducted values
    los=loss(y,h)           # loss call                 # losss call
    print('Loss :',los)
```
