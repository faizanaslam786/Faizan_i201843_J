```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("ECG200_TRAIN.csv",delimiter="  ",header=None)
df
```

```python
df=df.sort_values(by=[0],ascending = False)
df.index = range(len(df.index))
df
```

```python
df[0].value_counts()
```

## Making 70% of 69 and 30% of 31 for train

```python
x1train=df.iloc[:49]
# x1train
x1train_2=df.iloc[69:90]
# x1train_2
x1train=pd.concat([x1train, x1train_2])
```

```python
y1train=x1train[0]
x1train=x1train.drop([0], axis=1)
```

```python
# y1train[40:]
```

```python
x1test=df.iloc[49:69]
# x1test
x1test_2=df.iloc[90:]
# x1test_2
x1test=pd.concat([x1test, x1test_2])
```

```python
y1test=x1test[0]
x1test=x1test.drop([0], axis=1)
```

```python
# y1test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

clf=LogisticRegression()
clf.fit(x1train,y1train)

y_prediction=clf.predict(x1test)

own_acc = accuracy_score(y1test , y_prediction)
print("Accuracy by own splition :",own_acc)
```

## Using train test split fuction

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1train, y1train, test_size=0.3, random_state=42)

clf=LogisticRegression()
clf.fit(x_train,y_train)

y_prediction=clf.predict(x_test)

accuracy = accuracy_score(y_test , y_prediction)
print("Accuracy by builtin train test split model :",accuracy)
```
