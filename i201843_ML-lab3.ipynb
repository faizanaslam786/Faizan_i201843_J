# Task1-iii making projections on each other

```python
tmp=math.sqrt(np.dot(v1,v1.T))
proj2=(((np.dot(v1,v2))/tmp))*v1
print("Projection of v2 on v1 : ",proj2)

tmp2=math.sqrt(np.dot(v2,v2.T))
proj1=(((np.dot(v1,v2))/tmp2))*v2
print("Projection of v1 on v2 : ",proj1)
```

# Task1-v Findi eigenvalues and eigenvectors of iris data set

```python
df=pd.read_csv(f'iris.csv')
df.drop('variety' ,inplace=True, axis=1)
```

```python
df
```

```python
tv=df[['sepal.length', 'sepal.width','petal.length','petal.width']].to_numpy()
tv
```

```python
tv=np.asmatrix(tv)
tv=tv *tv.T
tv.shape
```

```python
tv.shape
```

```python
eigenvalues,eigenvectors = np.linalg.eig(tv)
print("Eigen vectors :",eigenvectors,'\n Eigen values',eigenvalues)
```
